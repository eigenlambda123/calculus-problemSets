### **Problems**

For 
$$ 
g(x) = \frac{x+7}{x^{2}-4} 
$$ 
evaluate,

(a) $\lim_{x \to 2^-} g(x)$  
(b) $\lim_{x \to 2^+} g(x)$  
(c) $\lim_{x \to 2} g(x)$  

---

### **Solutions**

For  
$$
g(x) = \frac{x+7}{x^{2}-4}=\frac{x+7}{(x-2)(x+2)}.
$$

#### (a) $\lim_{x \to 2^-} g(x)$

As $x\to 2^-$ we have $x<2$.

- Numerator: $x+7 \to 9$ (positive).  
- Factor $x+2 \to 4$ (positive).  
- Factor $x-2 \to 0^-$ (negative, since $x<2$).  
  Therefore the denominator $(x-2)(x+2)\to 0^-$.

A positive numerator divided by a very small negative denominator goes to $-\infty$:
$$
\lim_{x\to 2^-}\frac{x+7}{x^2-4} = -\infty.
$$

So,
$$
\boxed{\lim_{x \to 2^-} g(x) = -\infty}
$$


#### (b) $\lim_{x \to 2^+} g(x)$

As $x\to 2^+$ we have $x>2$.

- Numerator: $x+7 \to 9$ (positive).  
- Factor $x+2 \to 4$ (positive).  
- Factor $x-2 \to 0^+$ (positive, since $x>2$).  
  Therefore the denominator $(x-2)(x+2)\to 0^+$.

A positive numerator divided by a very small positive denominator goes to $+\infty$:
$$
\lim_{x\to 2^+}\frac{x+7}{x^2-4} = +\infty.
$$

So,
$$
\boxed{\lim_{x \to 2^+} g(x) = +\infty}
$$


#### (c) $\lim_{x \to 2} g(x)$

The one-sided limits are different:
$$
\lim_{x \to 2^-} g(x) = -\infty,\qquad
\lim_{x \to 2^+} g(x) = +\infty.
$$

Since the left and right limits are not equal, the two-sided limit does not exist (DNE):
$$
\boxed{\lim_{x \to 2} g(x)\ \text{does not exist (DNE).}}
$$

**Conclusion:** There is a vertical asymptote at $x=2$ (with sign $-\infty$ from the left and $+\infty$ from the right).
