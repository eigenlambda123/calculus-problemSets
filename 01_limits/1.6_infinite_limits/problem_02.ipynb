### **Problems**

For $$ h(t) = \frac{2t}{6+t} $$ evaluate,

(a) $\lim_{t \to -6^-} h(t)$ 
(b) $\lim_{t \to -6^+} h(t)$  
(c) $\lim_{t \to -6} h(t)$

---

### **Solutions**

For $h(t) = \dfrac{2t}{6+t}$.


#### (a) $\displaystyle \lim_{t \to -6^-} h(t)$

For $t \to -6^-$ we know $t < -6$.

- Numerator: $2t \to -12$ (negative).
- Denominator: $6+t \to 0$ and since $t < -6$, we also know $6+t < 0$.  
  Thus,
  $$
  6+t \to 0^-.
  $$

Therefore, a negative numerator divided by a very small **negative** denominator gives $+\infty$:
$$
\lim_{t \to -6^-} \frac{2t}{6+t} = +\infty
$$

So,
$$
\boxed{\lim_{t \to -6^-} h(t) = +\infty}
$$


#### (b) $\displaystyle \lim_{t \to -6^+} h(t)$

For $t \to -6^+$ we know $t > -6$.

- Numerator: $2t \to -12$ (negative).
- Denominator: $6+t \to 0$ and since $t > -6$, we have $6+t > 0$.  
  Thus,
  $$
  6+t \to 0^+.
  $$

Therefore, a negative numerator divided by a very small **positive** denominator gives $-\infty$:
$$
\lim_{t \to -6^+} \frac{2t}{6+t} = -\infty
$$

So,
$$
\boxed{\lim_{t \to -6^+} h(t) = -\infty}
$$


#### (c) $\displaystyle \lim_{t \to -6} h(t)$

From above,
$$
\lim_{t \to -6^-} h(t) = +\infty, 
\qquad
\lim_{t \to -6^+} h(t) = -\infty
$$

Since the one-sided limits are infinite and **do not agree**, the two-sided limit **does not exist**:
$$
\boxed{\lim_{t \to -6} h(t) \ \text{does not exist (DNE)}}
$$


**Conclusion:** There is a **vertical asymptote** at $t=-6$.
