### **Problems**

For $$ g(z) = \frac{z+3}{(z+1)^2} $$ evaluate,

(a) $\lim_{z \to -1^-} g(z)$
(b) $\lim_{z \to -1^+} g(z)$
(c) $\lim_{z \to -1} g(z)$

---

### **Solutions**

For  
$$
g(z) = \frac{z+3}{(z+1)^2}.
$$

#### (a) $\lim_{z \to -1^-} g(z)$

As $z \to -1^-$ we know $z < -1$.

- Numerator: $z+3 \to 2$ (positive).  
- Denominator: $(z+1)^2 \to 0$ but since it’s squared, it is **always positive**.  
  Thus,
  $$
  (z+1)^2 \to 0^+.
  $$

Therefore, positive numerator over very small positive denominator gives $+\infty$:
$$
\lim_{z \to -1^-} \frac{z+3}{(z+1)^2} = +\infty
$$

So,
$$
\boxed{\lim_{z \to -1^-} g(z) = +\infty}
$$


#### (b) $\lim_{z \to -1^+} g(z)$

As $z \to -1^+$ we know $z > -1$.

- Numerator: $z+3 \to 2$ (positive).  
- Denominator: $(z+1)^2 \to 0$ but again squared, so it’s **positive**.  
  Thus,
  $$
  (z+1)^2 \to 0^+.
  $$

Therefore, positive numerator over very small positive denominator gives $+\infty$:
$$
\lim_{z \to -1^+} \frac{z+3}{(z+1)^2} = +\infty
$$

So,
$$
\boxed{\lim_{z \to -1^+} g(z) = +\infty}
$$


#### (c) $\lim_{z \to -1} g(z)$

From above,
$$
\lim_{z \to -1^-} g(z) = +\infty, 
\qquad
\lim_{z \to -1^+} g(z) = +\infty
$$

Since both one-sided limits are $+\infty$, the two-sided limit is also $+\infty$:
$$
\boxed{\lim_{z \to -1} g(z) = +\infty}
$$

**Conclusion:** There is a **vertical asymptote** at $z = -1$.
