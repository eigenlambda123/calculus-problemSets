### **Problems**

For :$$ f(x) = \frac{9}{(x-3)^5} $$ evaluate,

(a) $\lim_{x \to 3^-} f(x)$
(b) $\lim_{x \to 3^+} f(x)$  
(c) $\lim_{x \to 3} f(x)$

---

### **Solutions**

For $f(x) = \dfrac{9}{(x-3)^5}$.

#### (a) $\displaystyle \lim_{x \to 3^-} f(x)$

For $x \to 3^-$ we know $x < 3$.

- Numerator: in the limit it is just $9$.
- Denominator: $x-3 \to 0$ and since $x<3$, we also know $x-3 < 0$.  
  Thus,
  $$
  x-3 \to 0^-.
  $$
  Raising to the fifth power preserves the sign (odd power), so
  $$
  (x-3)^5 \to 0^-.
  $$

Therefore, a positive numerator divided by a very small **negative** denominator gives $-\infty$:
$$
\lim_{x \to 3^-} \frac{9}{(x-3)^5} = -\infty
$$

So,
$$
\boxed{\lim_{x \to 3^-} f(x) = -\infty}
$$

#### (b) $\displaystyle \lim_{x \to 3^+} f(x)$

For $x \to 3^+$ we know $x > 3$.

- Numerator: $\to 9$.
- Denominator: $x-3 \to 0$ and since $x>3$, we have $x-3 > 0$.  
  Thus,
  $$
  x-3 \to 0^+ \quad\Rightarrow\quad (x-3)^5 \to 0^+.
  $$

Therefore, a positive numerator divided by a very small **positive** denominator gives $+\infty$:
$$
\lim_{x \to 3^+} \frac{9}{(x-3)^5} = +\infty
$$

So,
$$
\boxed{\lim_{x \to 3^+} f(x) = +\infty}
$$


#### (c) $\displaystyle \lim_{x \to 3} f(x)$

From above,
$$
\lim_{x \to 3^-} f(x) = -\infty, 
\qquad
\lim_{x \to 3^+} f(x) = +\infty
$$

Since the one-sided limits are infinite and **do not agree**, the two-sided limit **does not exist**:
$$
\boxed{\lim_{x \to 3} f(x) \ \text{does not exist (DNE)}}
$$


**Conclusion:** There is a **vertical asymptote** at $x=3$.
