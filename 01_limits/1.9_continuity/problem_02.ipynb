### **Problems**

The graph of $f(x)$ is given below. Based on this graph determine where the function is discontinuous.

### **Solutions**

Before starting, recall:  
For a function to be continuous at $x=a$, both $f(a)$ and $\lim_{x \to a} f(x)$ must exist, and we must have
$$
\lim_{x \to a} f(x) = f(a).
$$


#### At $x = -8$

- Left-hand limit:  
  $$
  \lim_{x \to -8^-} f(x) = -6
  $$
- Right-hand limit:  
  $$
  \lim_{x \to -8^+} f(x) = -6
  $$
- So,  
  $$
  \lim_{x \to -8} f(x) = -6
  $$
- However,  
  $$
  f(-8) = -3
  $$
- Since $\lim_{x \to -8} f(x) \neq f(-8)$, the function is not continuous.

Therefore,  
$$
\boxed{f(x)\ \text{is discontinuous at } x=-8}
$$  
because the limit and the function value differ.


#### At $x = -2$

- Left-hand limit:  
  $$
  \lim_{x \to -2^-} f(x) = 3
  $$
- Right-hand limit:  
  $$
  \lim_{x \to -2^+} f(x) = \infty
  $$
- Since left $\neq$ right, the two-sided limit does not exist.

Therefore,  
$$
\boxed{f(x)\ \text{is discontinuous at } x=-2}
$$  
because the limit does not exist.


#### At $x = 6$

- Left-hand limit:  
  $$
  \lim_{x \to 6^-} f(x) = 2
  $$
- Right-hand limit:  
  $$
  \lim_{x \to 6^+} f(x) = 5
  $$
- Since left $\neq$ right, the two-sided limit does not exist.

Therefore,  
$$
\boxed{f(x)\ \text{is discontinuous at } x=6}
$$  
because the limit does not exist.


#### **Final Answers**

The points of discontinuity are:  
$$
\boxed{x=-8,\quad x=-2,\quad x=6.}
$$
