### **Problems**

The graph of $f(x)$ is given below. Based on this graph determine where the function is discontinuous.

### **Solutions**

Before starting, recall:  
For a function to be continuous at $x=a$, both $f(a)$ and $\lim_{x \to a} f(x)$ must exist, and we must have
$$
\lim_{x \to a} f(x) = f(a).
$$


#### At $x = -4$

- Left-hand limit:  
  $$
  \lim_{x \to -4^-} f(x) = 3
  $$
- Right-hand limit:  
  $$
  \lim_{x \to -4^+} f(x) = -2
  $$
- Since the left- and right-hand limits are not equal, the two-sided limit does not exist.

Therefore,  
$$
\boxed{f(x)\ \text{is discontinuous at } x=-4}
$$  
because the limit does not exist.


#### At $x = 2$

- Left-hand limit:  
  $$
  \lim_{x \to 2^-} f(x) = -1
  $$
- Right-hand limit:  
  $$
  \lim_{x \to 2^+} f(x) = 5
  $$
- Again, left $\neq$ right, so the two-sided limit does not exist.

Therefore,  
$$
\boxed{f(x)\ \text{is discontinuous at } x=2}
$$  
because the limit does not exist.


#### At $x = 4$

- Left-hand and right-hand limits agree:  
  $$
  \lim_{x \to 4^-} f(x) = \lim_{x \to 4^+} f(x) = 2
  $$
- So,  
  $$
  \lim_{x \to 4} f(x) = 2
  $$
- However, $f(4)$ does not exist.

Therefore,  
$$
\boxed{f(x)\ \text{is discontinuous at } x=4}
$$  
because the function value is missing.


#### **Final Answers**

The points of discontinuity are:  
$$
\boxed{x=-4,\quad x=2,\quad x=4.}
$$
