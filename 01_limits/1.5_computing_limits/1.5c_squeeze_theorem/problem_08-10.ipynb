### **Problems**

**Evaluate each limit:**

8. **$\lim_{x \to 0} x \cos\left(\frac{1}{x}\right)$**

9. **$\lim_{x \to 0} x^2 \sin\left(\frac{1}{x^2}\right)$**

10. **$\lim_{x \to 0} x \sin\left(\frac{1}{x^2}\right)$**


---
    
### **Solutions**

**8.**  We want to evaluate  
$$
\lim_{x \to 0} x \cos\left(\frac{1}{x}\right).
$$  

Recall the cosine bound:  
$$
-1 \le \cos\left(\frac{1}{x}\right) \le 1
$$  

Multiply by $x$ (which can be positive or negative, but the inequality still holds in absolute value sense):  
$$
-x \le x \cos\left(\frac{1}{x}\right) \le x
$$  

Take limits as $x \to 0$:  
$$
\lim_{x \to 0} -x = 0, \quad \lim_{x \to 0} x = 0
$$  

By the Squeeze Theorem:  
$$
\lim_{x \to 0} x \cos\left(\frac{1}{x}\right) = 0
$$  

$$
\boxed{0}
$$


**9.**  We want to evaluate  
$$
\lim_{x \to 0} x^2 \sin\left(\frac{1}{x^2}\right).
$$  

Recall the sine bound:  
$$
-1 \le \sin\left(\frac{1}{x^2}\right) \le 1
$$  

Multiply by $x^2 \ge 0$:  
$$
-x^2 \le x^2 \sin\left(\frac{1}{x^2}\right) \le x^2
$$  

Take limits as $x \to 0$:  
$$
\lim_{x \to 0} -x^2 = 0, \quad \lim_{x \to 0} x^2 = 0
$$  

By the Squeeze Theorem:  
$$
\lim_{x \to 0} x^2 \sin\left(\frac{1}{x^2}\right) = 0
$$  

$$
\boxed{0}
$$


**10.**  We want to evaluate  
$$
\lim_{x \to 0} x \sin\left(\frac{1}{x^2}\right).

$$
-1 \le \sin\left(\frac{1}{x^2}\right) \le 1
$$  

Multiply by $x$:

-x \le x \sin\left(\frac{1}{x^2}\right) \le x
$$  

Take limits as $x \to 0$:  
$$
\lim_{x \to 0} -x = 0, \quad \lim_{x \to 0} x = 0
$$  

By the Squeeze Theorem:  
$$
\lim_{x \to 0} x \sin\left(\frac{1}{x^2}\right) = 0
$$  

$$
\boxed{0}
$$
