### **Problem**

Let $f, g,$ and $h$ be the functions defined by  

$$
f(x) = \frac{1 - \cos x}{2x}, \quad 
g(x) = x^3 \sin\!\left(\frac{1}{x}\right), \quad 
h(x) = \frac{x}{\sin x}, \quad \text{for } x \neq 0.
$$  

Determine whether the following inequalities can be used with the Squeeze Theorem to find the limit of the function as $x \to 0$.

---

### **Solutions**

**4.**  $-\frac{1}{2} \le f(x) \le x^2 + \frac{1}{2}$ 

We want to evaluate  
$$
\lim_{x \to 0} f(x).
$$  

The bounds are  
$$
\lim_{x \to 0} -\frac{1}{2} = -\frac{1}{2}, \quad 
\lim_{x \to 0} \left(x^2 + \frac{1}{2}\right) = \frac{1}{2}.
$$  

Since the bounds do not converge to the same value, the Squeeze Theorem **cannot** be used.  

$$
\boxed{\text{Cannot use Squeeze Theorem}}
$$


**5.**  $x^3 \le g(x) \le x^3$  

We want to evaluate  
$$
\lim_{x \to 0} g(x).
$$  

The bounds are equal:  
$$
\lim_{x \to 0} x^3 = 0
$$  

Since both bounds converge to the same limit, the Squeeze Theorem **can** be used:  
$$
\lim_{x \to 0} g(x) = 0
$$  

$$
\boxed{0}
$$


**6.**  $1 - x^2 \le h(x) \le x^2 + 1$  

We want to evaluate  
$$
\lim_{x \to 0} h(x).
$$  

The bounds converge to the same limit:  
$$
\lim_{x \to 0} (1 - x^2) = 1, \quad 
\lim_{x \to 0} (x^2 + 1) = 1
$$  

By the Squeeze Theorem:  
$$
\lim_{x \to 0} h(x) = 1
$$

$$
\boxed{1}
$$
