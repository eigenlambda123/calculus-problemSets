### **Problem**

- Compute the following limit using limit properties. At each step, clearly indicate the property being used. If not possible to compute, explain why.

$$
\lim_{w \to 3} \left( \frac{w^{2} - 8w}{4 - 7w} \right)
$$

---



### **Solution**

1. **Quotient Rule**:  

The limit of a quotient is the quotient of the limits, provided the denominatorâ€™s limit is not zero:  
$$
\lim_{w \to 3} \frac{w^{2} - 8w}{4 - 7w}
= \frac{\lim_{w \to 3} (w^{2} - 8w)}{\lim_{w \to 3} (4 - 7w)}
$$

2. **Sum/Difference Rule (Numerator)**:  
$$
\lim_{w \to 3} (w^{2} - 8w) 
= \lim_{w \to 3} w^{2} - \lim_{w \to 3} 8w
$$

3. **Constant Multiple Rule (Numerator)**:  
$$
\lim_{w \to 3} 8w = 8 \cdot \lim_{w \to 3} w
$$

4. **Direct Substitution (Numerator)**:  
$$
\lim_{w \to 3} w^{2} = (3)^{2} = 9
$$  
$$
\lim_{w \to 3} w = 3
$$  
Therefore:  
$$
\lim_{w \to 3} (w^{2} - 8w) = 9 - 8(3) = 9 - 24 = -15
$$

5. **Sum/Difference Rule (Denominator)**:  
$$
\lim_{w \to 3} (4 - 7w) 
= \lim_{w \to 3} 4 - \lim_{w \to 3} 7w
$$

6. **Constant Multiple Rule (Denominator)**:  
$$
\lim_{w \to 3} 7w = 7 \cdot \lim_{w \to 3} w
$$

7. **Direct Substitution (Denominator)**:  
$$
\lim_{w \to 3} 4 = 4
$$  
$$
\lim_{w \to 3} w = 3
$$  
Therefore:  
$$
\lim_{w \to 3} (4 - 7w) = 4 - 7(3) = 4 - 21 = -17
$$

8. **Compute the Quotient**:  
$$
\frac{\lim_{w \to 3} (w^{2} - 8w)}{\lim_{w \to 3} (4 - 7w)}
= \frac{-15}{-17}
$$  
$$
= \frac{15}{17}
$$

$$
\boxed{\frac{15}{17}}
$$
