### **Problem**

- Compute the following limit using limit properties. At each step, clearly indicate the property being used. If not possible to compute, explain why.

$$
\lim_{x \to -5} \left( \frac{x+7}{x^{2} + 3x - 10} \right)
$$

---


### **Solution**

1. **Quotient Rule**:  

The limit of a quotient is the quotient of the limits, provided the denominator’s limit is not zero:  
$$
\lim_{x \to -5} \frac{x+7}{x^{2} + 3x - 10}
= \frac{\lim_{x \to -5} (x+7)}{\lim_{x \to -5} (x^{2} + 3x - 10)}
$$

2. **Sum/Difference Rule (Numerator)**:  

$$
\lim_{x \to -5} (x+7) = \lim_{x \to -5} x + \lim_{x \to -5} 7
$$

3. **Constant Rule & Direct Substitution (Numerator)**:  

$$
\lim_{x \to -5} x = -5
$$  
$$
\lim_{x \to -5} 7 = 7
$$  
Therefore:  
$$
\lim_{x \to -5} (x+7) = -5 + 7 = 2
$$

4. **Sum/Difference Rule (Denominator)**:  

$$
\lim_{x \to -5} (x^{2} + 3x - 10)
= \lim_{x \to -5} x^{2} + \lim_{x \to -5} 3x - \lim_{x \to -5} 10
$$

5. **Constant Multiple Rule & Direct Substitution (Denominator)**:  

$$
\lim_{x \to -5} x^{2} = (-5)^{2} = 25
$$  
$$
\lim_{x \to -5} 3x = 3 \cdot (-5) = -15
$$  
$$
\lim_{x \to -5} 10 = 10
$$  
Therefore:  
$$
\lim_{x \to -5} (x^{2} + 3x - 10) = 25 - 15 - 10 = 0
$$

6. **Check the Denominator**:  

Since the denominator’s limit is \(0\) and the numerator’s limit is \(2 \neq 0\), the limit does **not** exist as a finite number — the function goes to infinity or negative infinity depending on direction.

7. **Conclusion**:  

The two-sided limit does not exist in the finite sense because:  
$$
\frac{\text{nonzero number}}{0} \quad \Rightarrow \quad \text{infinite limit or divergence}.
$$ 

$$
\boxed{\text{Limit does not exist (infinite divergence)}}
$$
